APPLYING CONSTRAINTS ON USERS AND BOOKS
USERS-ONLY THOSE USERS WHO HAVE RATED ATLEAST 200 BOOKS
BOOKS- ONLY THOSE BOOKS WHICH HAVE BEEN RATED BY ATLEAST 50 USERS

```python
x=ratings['User-ID'].value_counts()>200
```

```python
ratings=ratings[ratings['User-ID'].isin(y)]
```

```python
rating_withbooks=ratings.merge(books, on='ISBN')
rating_withbooks
```

```python
rating_withbooks.shape
```

```python
number_rating=rating_withbooks.groupby('Title')['Book-Rating'].count().reset_index()
number_rating
```

```python
number_rating.rename(columns={'Book-Rating':'number of ratings'},inplace=True)
number_rating
```

```python
final_rating=number_rating.merge(rating_withbooks,on='Title')
final_rating
```

```python
final_rating=final_rating[final_rating['number of ratings']>50]
final_rating
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='User-ID',index='Title',values='Book-Rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[240,:].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestions
```

```python
distances
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])


```

```python
book_pivot.index[240]
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
  for i in range(len(suggestions)):
    if i==0:
      print("the suggestions for" ,book_name, "are:")
    if not i:
      print(book_pivot.index[suggestions[i]])




```

```python
recommend_book('Hideaway')
```
